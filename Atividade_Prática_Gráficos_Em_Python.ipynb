```python
import pandas as pd

# Dados
Redes_Sociais = ['Whatsapp', 'Youtube', 'Instagram', 'Facebook', 'Tiktok', 'LinkedIn', 'Messenger']
Quantidade = [169000000, 142000000, 113000000, 109000000, 82000000, 63000000, 62000000]

# Criar um DataFrame
data = {
    'Rede_Social': Redes_Sociais,
    'Quantidade': Quantidade
}
df = pd.DataFrame(data)

# Salvar DataFrame em um arquivo CSV
df.to_csv('dados.csv', index=False)

print("Arquivo CSV 'dados.csv' criado com sucesso!")
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Carregar dataset a partir do arquivo CSV
df = pd.read_csv('dados.csv')

# Extrair as colunas do DataFrame
Redes_Sociais = df['Rede_Social'].tolist()
Quantidade = df['Quantidade'].tolist()
tempo = range(len(Redes_Sociais))

# Verificar os estilos disponíveis
print(plt.style.available)

# Usar o estilo seaborn-v0_8
plt.style.use('seaborn-v0_8')

# Configurar a grade
plt.grid(True, zorder=0)

# Cores para as barras
cores_hex = ['#007800', '#c92c2c', '#962c4c', '#101652', '#140d1a', '#00aaff', '#207178']

# Criar gráfico de barras
plt.bar(Redes_Sociais, Quantidade, color=cores_hex, zorder=2)

# Converter Quantidade para milhares
Quantidade_em_milhares = [num / 1000 for num in Quantidade]

# Criar gráfico de linha sobreposto
plt.plot(tempo, Quantidade_em_milhares, marker='o', linestyle='-', color='black', zorder=3)

# Configurar rótulos e títulos
plt.xlabel('Redes Sociais', fontsize=11)
plt.ylabel('Quantidade de Usuários (em milhares)', fontsize=12)
plt.title('Quantidade de Usuários', fontsize=13, loc='center')
plt.suptitle('Dados referentes ao ano de 2023 no Brasil', fontsize=15, ha='center')

# Ajustar rotação dos rótulos do eixo x
plt.xticks(rotation=45)

# Configurar a família da fonte
plt.rcParams.update({'font.family': 'serif', 'font.serif': ['Times New Roman']})

# Ajustar layout
plt.tight_layout(rect=[0.04, 0.1, 0.9, 0.9])
plt.subplots_adjust(top=0.86)

# Mostrar gráfico
plt.show()
```
